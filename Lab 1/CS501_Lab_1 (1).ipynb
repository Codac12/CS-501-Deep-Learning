```
import IPython.display
import PIL.Image
import numpy as np
import matplotlib.pyplot as plt


from PIL import Image, ImageDraw

for i in range(5):
  red = np.random.randint(1,256)
  green = np.random.randint(1,256)
  blue = np.random.randint(1,256)
  im = Image.new('RGB', (512, 288), (red,green,blue))
  draw = ImageDraw.Draw(im)
  
  for i in range(100):
    red = np.random.randint(1,256)
    green = np.random.randint(1,256)
    blue = np.random.randint(1,256)
    start_angle = np.random.randint(1,360)
    end_angle = np.random.randint(1,360)
    start_x = np.random.randint(1,512)
    start_y = np.random.randint(1,288)
    end_x = np.random.randint(1,512)
    end_y = np.random.randint(1,288)
    
    draw.arc([(start_x^2, start_y), (end_x, end_y)],start_angle,end_angle,(red,green,blue))
  for j in range(20):
    red = np.random.randint(1,256)
    green = np.random.randint(1,256)
    blue = np.random.randint(1,256)
    start_angle = np.random.randint(1,360)
    end_angle = np.random.randint(1,360)
    start_x = np.random.randint(1,512)
    start_y = np.random.randint(1,288)
    end_x = np.random.randint(1,512)
    end_y = np.random.randint(1,288)
    width = np.random.randint(1,10)
    draw.line([(start_x, start_y), (end_x, end_y)],(red,green,blue), width)

    
    

  plt.imshow(np.asarray(im))
 
```
